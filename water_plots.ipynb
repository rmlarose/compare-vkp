```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
fci_energy = -7.572938704e+01
```

```python
f = h5py.File("data/water_u.hdf5", "r")
ds = f["ds"][:]
energies = f["energies"][:]
f.close()
```

```python
fig, ax = plt.subplots()
ax.plot(ds, energies, '.')
ax.hlines(fci_energy, 0., np.max(ds))
```
