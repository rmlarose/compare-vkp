```python
import h5py
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt
```

```python
f1 = h5py.File("lih_scratch/subspace.hdf5", "r")
my_h = f1["h"][:, :]
my_s = f1["s"][:, :]
f1.close()

print(my_h.shape)
d_max = my_h.shape[0]
```

```python
f2 = h5py.File("lih_scratch_new/subspace.hdf5", "r")
hpcc_h = f2["h"][:, :]
hpcc_s = f2["s"][:, :]
f2.close()

print(hpcc_h.shape)

hpcc_h_sub = hpcc_h[:d_max, :d_max]
hpcc_s_sub = hpcc_s[:d_max, :d_max]
print(hpcc_h_sub.shape)
```

```python
h_err = np.abs(hpcc_h_sub - my_h)
s_err = np.abs(hpcc_s_sub - my_s)
```

```python
fig, ax = plt.subplots(2, 1)
im1 = ax[0].imshow(np.abs(h_err))
im2 = ax[1].imshow(np.abs(s_err))
fig.colorbar(im1)
fig.colorbar(im2)
```
