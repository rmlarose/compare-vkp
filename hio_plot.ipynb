```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
f = h5py.File("data/phases.hdf5", "r")
amplitdues = f["amplitudes"][:, :]
real_F = f["F_real"][:, :]
amplitudes_hio = f["amplitudes_hio"][:, :]
deltas = f["delta"][:]
f.close()
```

```python
phases = np.angle(amplitdues)
```

```python
fig, ax = plt.subplots()
plt.grid(False)
im = ax.imshow(phases, cmap='plasma')
fig.colorbar(im)
```

```python
fig2, ax2 = plt.subplots(2, 1)
plt.grid(False)
im = ax2[0].imshow(np.abs(real_F), cmap='plasma')
im = ax2[1].imshow(np.angle(real_F), cmap='plasma')
fig2.colorbar(im)
```

```python
phases_hio = np.angle(amplitudes_hio)

fig3, ax3 = plt.subplots()
plt.grid(False)
im = ax3.imshow(phases_hio, cmap='plasma')
fig3.colorbar(im)
```

```python
err = np.abs(phases - phases_hio)

fig4, ax4 = plt.subplots()
plt.grid(False)
im = ax4.imshow(err, cmap='plasma')
fig4.colorbar(im)
```
