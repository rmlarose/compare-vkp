```python
import h5py
import pandas as pd
import matplotlib.pyplot as plt
```

```python
output_file = "water_data_new/eigvals.hdf5"
df = pd.read_hdf(output_file, key="eigenvalues")
```

```python
print(df.head())
```

```python
output_file2 = "water_10steps/eigvals.hdf5"
df2 = pd.read_hdf(output_file2, key="eigenvalues")
```

```python
energy_dmrg = -75.72977583
energy_fci = -7.572938704e+01
```

```python
fig, ax = plt.subplots()
ax.plot(df["d"], df["energy"], label="Krylov 20 steps")
ax.plot(df2["d"], df2["energy"], label="Krylov 10 steps")
ax.hlines(energy_dmrg, 0., max(df["d"]), 'k', label="DMRG")
ax.hlines(energy_fci, 0., max(df["d"]), 'g', label="FCI")

x1, x2, y1, y2 = 7, 10, -75.74, -75.72
ax_ins = ax.inset_axes(
    [0.2, 0.4, 0.3, 0.3],
    xlim=(x1, x2), ylim=(y1, y2)
)
ax_ins.plot(df["d"], df["energy"], label="Krylov 20 steps")
ax_ins.plot(df2["d"], df2["energy"], label="Krylov 10 steps")
ax_ins.hlines(energy_dmrg, 0., max(df["d"]), 'k', label="DMRG")
ax_ins.hlines(energy_fci, 0., max(df["d"]), 'g', label="FCI")

plt.legend()
```
